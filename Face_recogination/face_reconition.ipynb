```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread ("../input/face-recognition/group.jpg")
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
haar_face_cascade = cv2.CascadeClassifier("../input/face-recognition/haarcascade_frontalface_default.xml")
```

```python
faces = haar_face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)
```

```python
for x,y,w,h in faces:
    cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,0),3)
```

```python
print("Faces Found", len(faces))
```

```python
plt.rcParams["figure.figsize"] = (14,10)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
#plt.show()
plt.savefig("face_detection.png", dpi=400, bbox_inches='tight')
```
